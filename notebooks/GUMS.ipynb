# Working with the Gaia Universe Model Snapshot - 10th version

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table

%matplotlib inline
```

```python
filename = "../data/GUMS/gum_mw.sam"
readme = "../data/GUMS/ReadMe"
GUMS = Table.read(filename, format="cds", readme=readme)
```

```python
print GUMS.dtype.names
print len(GUMS)
```

```python
plt.hist(GUMS["Mass"], bins=20, histtype='step', color='k')

plt.xlabel(r'${\rm Mass}\ (M_{\odot})$', size=16)
plt.ylabel(r"${\rm N}$", size=16)
plt.show()
```

```python
plt.scatter(GUMS["RA2kdeg"], GUMS["DE2kdeg"], marker='.')

plt.show()
```
